```python
music_years = music[music['song.year']!=0].groupby('song.year')[['song.tempo', 'song.loudness']].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(data = music_years['song.tempo'], color='g');
ax2 = plt.twinx()
sns.lineplot(data = music_years['song.loudness'], color='b', ax=ax2);
```

```python
ax = music_years.plot(x='song.year', y='song.tempo', legend=False)
ax2 = plt.twinx()
music_years.plot(x='song.year', y='song.loudness', ax=ax2, legend=False, color='r')
ax.figure.legend()
plt.show();
```

```python
ax = music_years.loc['1960':].plot(x='song.year', y='song.tempo', legend=False, figsize=(12,7))
plt.title('Songs mean tempo and loudness trend', {'size':20})
plt.ylabel('tempo', {'size':18}, color='b')
plt.xlabel('year',{'size':18})
ax2 = plt.twinx()
ax2.set_ylabel('loudness',  {'size':18}, color='r')
music_years.loc['1960':].plot(x='song.year', y='song.loudness', ax=ax2, legend=False, color='r')
ax.figure.legend()
plt.show();
```
